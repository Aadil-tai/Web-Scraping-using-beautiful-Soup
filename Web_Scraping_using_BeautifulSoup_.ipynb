## <font color='green'> Searching - find() and find_all()</font>

When scraping web pages, it's often necessary to locate specific elements within the HTML structure. BeautifulSoup provides two primary methods for this purpose: `find()` and `find_all()`.

### Using find()

The `find()` method is used to locate the first occurrence of a specific tag or set of tags. It returns only the first match found in the HTML document.

## <font color="Green"> Processing multiple links at once

```python
[l.get('href')for l in links]
```

- I extracted all the link destinations (URLs) from a list of HTML links using the code `[l.get('href') for l in links]`.

- I did this because I needed a list of all the URLs that the links in the HTML pointed to. This could be useful for various tasks, such as web scraping or analyzing the link structure of a website.

- Now with Proper Indexing

```python
# Assuming 'links' is your list of link elements
for i, link in enumerate(links):
    href = link.get('href')
    if href:  # Check if href exists
        print(f"{i+1}. {href}")
```

```python
clean_links = [l for l in links if l.get('href') != None]
```

- I created a new list called `clean_links` using the code `[l for l in links if l.get('href') != None]`.

- I did this to filter out any links that don't have an `href` attribute (meaning they don't actually point to a URL). This ensures that the resulting list only contains valid links.

```python
relative_links = [l.get('href') for l in clean_links]
```

```python
# Assuming 'relative_links' is your list of relative URLs
for i, url in enumerate(relative_links):
    print(f"{i+1}. {url}")
```

```python
relative_links
```

- I extracted the relative URLs from the `clean_links` list using the code `relative_links = [l.get('href') for l in clean_links]`.

- I did this because I specifically wanted the relative URLs, which are the parts of the URLs that don't include the domain name. This is useful when you want to focus on the internal structure of a website, or when you need to construct full URLs later by combining the relative URLs with the base URL of the website.

```python
full_urls = [urljoin(base_site,url) for url in relative_links]
```

```python
# Assuming 'full_urls' is your list of URLs
for i, url in enumerate(full_urls):
    print(f"{i+1}. {url}")
```

- I created a list of full URLs using the code `full_urls = [urljoin(base_site,url) for url in relative_links]`.

- I did this to construct the complete URLs by combining the base URL of the website (`base_site`) with the relative URLs I had extracted earlier. This gives me a list of URLs that can be directly accessed.

```python
titles = []
for link in links:
    if link.has_attr('href'):
        title = link.get('title')
        titles.append(title)

# Clean the list from None values
clean_titles = [t for t in titles if t is not None]

# Printing titles in respective lines with numbers
for i, title in enumerate(clean_titles):
    print(f"{i+1}. {title}")
```

```python
import pandas as pd
import os

# Assuming 'notebook_directory' is already defined and contains the path to your notebook directory
contents_folder = os.path.join(Notebook_directory, 'contents')

# Create the 'contents' folder if it doesn't exist
if not os.path.exists(contents_folder):
    os.makedirs(contents_folder)

output_file = os.path.join(contents_folder, "extracted_links.xlsx")

# Check if the file already exists
if not os.path.exists(output_file):
    # Create a list of dictionaries where each dictionary contains the index and full URL of a link.
    data = []
    for i, link in enumerate(full_urls):
        data.append({
            "Index": i + 1,
            "Full URL": link
        })

    # Create a Pandas DataFrame from the list of dictionaries.
    df = pd.DataFrame(data)

    # Export the DataFrame to an Excel file within the 'contents' folder.
    df.to_excel(output_file, sheet_name="Sheet1", index=False)
    print(f"Excel file saved to: {output_file}")
else:
    print(f"File '{output_file}' already exists. Skipping creation.")
```
